```python
def compute_metrics_with_travel(events):
    """
    events : list of tuples (day_num, hour, campus, duration_minutes)
    Returns dict with keys:
        late (events at hour == 8),
        lunch (events at hour in {3,4}),
        isolated (hour 1..7 with event and no event at h±1, same day),
        days (number of distinct days with any event),
        wednesday (events on Wednesday (day 2) with hour >= 4),
        travel (total travel minutes between consecutive events on same day)
    """
    if not events:
        return {'late':0, 'lunch':0, 'isolated':0, 'days':0, 'wednesday':0, 'travel':0}
    
    # Sort by day then hour
    events.sort(key=lambda x: (x[0], x[1]))
    
    event_set = {(d, h) for d, h, _, _ in events}
    late = 0
    lunch = 0
    wednesday = 0
    days_set = set()
    travel = 0
    
    # First pass: count simple metrics and collect days
    for d, h, _, _ in events:
        days_set.add(d)
        if h == 8:
            late += 1
        if h in (3,4):
            lunch += 1
        if d == 2 and h >= 4:
            wednesday += 1
    
    # Isolated hours: h=1..7, event present, no event at h-1 or h+1 on same day
    isolated = 0
    for d, h, _, _ in events:
        if 1 <= h <= 7:
            if (d, h-1) not in event_set and (d, h+1) not in event_set:
                isolated += 1
    
    # Travel between consecutive events on the same day
    for i in range(len(events)-1):
        d1, h1, camp1, dur1 = events[i]
        d2, h2, camp2, dur2 = events[i+1]
        if d1 == d2:
            # travel time between campuses
            key = (camp1, camp2)
            travel += travel_time_dict.get(key, 60)   # default 60 min if unknown
            # Gap not used – we skip it as requested
    return {
        'late': late,
        'lunch': lunch,
        'isolated': isolated,
        'days': len(days_set),
        'wednesday': wednesday,
        'travel': travel
    }
```

```python
# Apply Monday-Friday filter (days 0-4) to sem1
sem1 = sem1[sem1['DayNum'].isin([0,1,2,3,4])].copy()
# -------------------------------------------------------------------
# Build module_event_map (one entry per base module)
# -------------------------------------------------------------------
module_events = {}
for _, row in sem1.iterrows():
    base = row['ModCode']
    event = (row['DayNum'], row['ModelHour'], row['Campus'], row['Duration (minutes)'])
    module_events.setdefault(base, []).append(event)
# Deduplicate events per module
for base in module_events:
    module_events[base] = list(set(module_events[base]))

# -------------------------------------------------------------------
# Helper: check if candidate events clash with baseline events
# -------------------------------------------------------------------
def has_clash(baseline_events, cand_events):
    baseline_set = {(d, h) for d, h, _, _ in baseline_events}
    for d, h, _, _ in cand_events:
        if (d, h) in baseline_set:
            return True
    return False

# -------------------------------------------------------------------
# Build estimation dataset using bundle replacement
# -------------------------------------------------------------------
estimation_rows = []

for student_id in sem1['AnonID'].unique():
    student_df = sem1[sem1['AnonID'] == student_id]
    prog_base = student_df['ProgBase'].iloc[0]
    
    # Compulsory modules (unique)
    comp_bases = set(student_df[student_df['Compulsory'] == True]['ModCode'])
    comp_events = []
    for base in comp_bases:
        comp_events.extend(module_events.get(base, []))
    
    # Chosen optional modules (unique)
    chosen_bases = set(student_df[student_df['Compulsory'] == False]['ModCode']) & set(module_events.keys())    
    if not chosen_bases:
        continue
    
    # Universal optional modules for this programme: those marked optional in pc_agg,
    # plus the chosen modules (since they are obviously available).
    optional_for_prog = set(pc_agg[(pc_agg['ProgBase'] == prog_base) & (pc_agg['Compulsory'] == False)]['ModCode'])
    optional_universe = optional_for_prog & set(module_events.keys())
    optional_universe.update(chosen_bases)   # ensure chosen modules are included
    # Diagnostic
    print(f"Student {student_id}: comp_bases {comp_bases}")
    print(f"chosen_bases {chosen_bases}")
    print(f"optional_for_prog size {len(optional_for_prog)}")
    print(f"sem1_base_modules size {len(set(module_events.keys()))}")
    print(f"optional_universe size {len(optional_universe)}")
    if len(optional_universe) == 0:
        print("optional_universe is empty – check pc_agg and module_events keys")
        continue   # skip this student
    print("Number of module_events keys:", len(module_events))
    
    sample_student = sem1['AnonID'].iloc[0]
    sd = sem1[sem1['AnonID'] == sample_student]
    
    print("Sample student:", sample_student)
    print("Chosen ModCodes:", set(sd[sd['Compulsory'] == False]['ModCode']))
    
    for base in set(sd[sd['Compulsory'] == False]['ModCode']):
        print(base, "events:", len(module_events.get(base, [])))
    for chosen_base in chosen_bases:
        # Baseline schedule: compulsory + all other chosen optionals (exclude chosen_base)
        other_optional_bases = chosen_bases - {chosen_base}
        baseline_events = comp_events.copy()
        for base in other_optional_bases:
            baseline_events.extend(module_events.get(base, []))
        baseline = compute_metrics_with_travel(baseline_events)
        
        # Candidate modules: from optional_universe that are not already in chosen_bases
        # (the chosen_base itself is the "chosen" alternative)
        candidate_bases = list(optional_universe - chosen_bases)
        # if len(candidate_bases) > K:
        #     candidate_bases = np.random.choice(candidate_bases, size=K, replace=False).tolist()
        # else:
        #     candidate_bases = list(candidate_bases)
        
        all_candidates = candidate_bases + [chosen_base]
        # The chosen module is one of the candidates (we'll add it separately)
        
        for cand_base in all_candidates:
            cand_events = module_events.get(cand_base, [])
            if not cand_events:
                continue
            # Skip if candidate (other than the chosen one) clashes with baseline schedule
            if cand_base != chosen_base and has_clash(baseline_events, cand_events):
                continue
            combined_events = baseline_events + cand_events
            combined = compute_metrics_with_travel(combined_events)
            
            inc_late = combined['late'] - baseline['late']
            inc_lunch = combined['lunch'] - baseline['lunch']
            inc_isolated = combined['isolated'] - baseline['isolated']
            inc_days = combined['days'] - baseline['days']
            inc_wednesday = combined['wednesday'] - baseline['wednesday']
            inc_travel = combined['travel'] - baseline['travel']
            
            # Scale travel to keep magnitude reasonable
            # inc_travel /= 10
            
            chosen_flag = 1 if cand_base == chosen_base else 0
            if cand_base == chosen_base:
                print(f"Adding row for student {student_id}, chosen module {cand_base}")
            estimation_rows.append({
                'student_id': student_id,
                'module': cand_base,
                'chosen': chosen_flag,
                'inc_late': inc_late,
                'inc_lunch': inc_lunch,
                'inc_isolated': inc_isolated,
                'inc_days': inc_days,
                'inc_wednesday': inc_wednesday,
                'inc_travel': inc_travel
            })

choice_data = pd.DataFrame(estimation_rows)
print(f"Choice dataset shape: {choice_data.shape}")
print(choice_data.head())
```

```python
# -------------------------------------------------------------------
# Logistic regression to estimate weights (including travel)
# -------------------------------------------------------------------
import statsmodels.api as sm

X = choice_data[['inc_late', 'inc_lunch', 'inc_isolated', 'inc_days', 'inc_wednesday', 'inc_travel']]
X = sm.add_constant(X)
y = choice_data['chosen']

logit_model = sm.Logit(y, X)
result = logit_model.fit()
print(result.summary())
coeffs = result.params
print("\nEstimated coefficients (utility weights):")
print(coeffs)
```

```python
# After building `merged` (which contains all events with week numbers), 
# expand the data so each event row is repeated for each week it occurs.
def parse_weeks(weeks_str):
    if pd.isna(weeks_str):
        return []
    weeks = []
    for part in str(weeks_str).split(','):
        if '-' in part:
            start, end = map(int, part.split('-'))
            weeks.extend(range(start, end + 1))
        else:
            weeks.append(int(part))
    return weeks

expanded_rows = []
for _, row in merged.iterrows():
    weeks = parse_weeks(row['Weeks'])
    for w in weeks:
        new_row = row.copy()
        new_row['Week'] = w
        expanded_rows.append(new_row)
expanded_df = pd.DataFrame(expanded_rows)

# Now group by student and week to build the schedule for that week
student_weekly_travel = []
for student_id in expanded_df['AnonID'].unique():
    student_df = expanded_df[expanded_df['AnonID'] == student_id]
    weeks = student_df['Week'].unique()
    for w in weeks:
        week_df = student_df[student_df['Week'] == w]
        # Build list of events for this week: (day_num, model_hour, campus, duration)
        events = []
        for _, row in week_df.iterrows():
            events.append((row['DayNum'], row['ModelHour'], row['Campus'], row['Duration (minutes)']))
        if events:
            metrics = compute_metrics_with_travel(events)
            student_weekly_travel.append(metrics['travel'])
    # Print progress (optional)
    # if len(student_weekly_travel) % 100 == 0:
    #     print(f"Processed {len(student_weekly_travel)} weeks")

# Now we have a list of weekly travel times for all students
avg_travel = np.mean(student_weekly_travel)
max_travel = np.max(student_weekly_travel)
print(f"Average weekly travel time: {avg_travel:.2f} minutes")
print(f"Maximum weekly travel time: {max_travel:.2f} minutes")
```
